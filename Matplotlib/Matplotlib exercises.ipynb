```python
theta = np.linspace(0,2*np.pi,1000)
y = np.sin(theta)
x = np.cos(theta)
plt.plot(x, y,color = '#3399ff',linewidth=3)
plt.axis('equal')
plt.fill(x,y,color = '#3399ff',alpha=0.2)
plt.plot(0,0,'ko')
plt.text(0,0,'(0,0)',horizontalalignment = 'left',verticalalignment = 'top', fontsize = 12)
plt.show()
```

```python
import datetime
now = datetime.datetime.now()
```

```python
hour_clc = -(now.hour/12+now.minute/60/12) * 2*np.pi + np.pi*0.5
hour_clc_x = np.cos(hour_clc)
hour_clc_y = np.sin(hour_clc)
```

```python
minute_clc = -(now.minute/60) * 2*np.pi + np.pi*0.5
minute_clc_x = np.cos(minute_clc)
minute_clc_y = np.sin(minute_clc)
```

```python
theta = np.linspace(0,2*np.pi,1000)
y = np.sin(theta)
x = np.cos(theta)
plt.plot(x, y,color = '#fff2cc',linewidth=3)
plt.axis('equal')
plt.fill(x,y,color = '#fff2cc',alpha=0.3)
plt.plot(0,0,'ko')
plt.text(0,0,'',horizontalalignment = 'left',verticalalignment = 'top', fontsize = 12)

hours_signs = np.r_[3,2,1,np.arange(12,3,-1)]
for idx in range(len(hours_x)):
    plt.text(hours_x[idx] - hours_x[idx]*0.1 ,hours_y[idx] - hours_y[idx]*0.1,hours_signs[idx],horizontalalignment = 'center',verticalalignment = 'center'
            ,color ='#804000')

plt.plot([0,minute_clc_x*0.8],[0,minute_clc_y*0.8],color = '#663300')
plt.plot([0,hour_clc_x*0.6],[0,hour_clc_y*0.6],linewidth=3,color ='#663300')
plt.show()
```

```python
flights = sns.load_dataset("flights")
sredniPasazerowie = flights["passengers"].groupby(flights["year"]).agg(np.mean)
sredniPasazerowie

fig,ax = plt.subplots()
ax.plot(sredniPasazerowie,ls='--',lw=3)
ax.plot(sredniPasazerowie,'bo')
ax.set_title('Wykres średniej liczby pasażerów',color = 'k',fontsize =14)
ax.set_ylabel('Liczba pasażerów',color ='b')
ax.set_xlabel('Rok',color ='b')
ax.axhline(y=240,color='#5c00e6',ls='--',alpha=0.5)
ax.axhline(y=300,color='#5c00e6',ls='--',alpha=0.5)
ax.axvline(x=1956,color='#5c00e6',ls='--',alpha=0.5)
ax.axvline(x=1952,color='#5c00e6',ls='--',alpha=0.5)
ax.grid(True,linewidth=0.5,color = '#cccccc',linestyle='--')
```

```python
cmap = plt.cm.get_cmap('BuPu')
colors = np.random.choice(1000,2,replace=False)/1000

tips = sns.load_dataset('tips')
smokers = tips.groupby('smoker').size()
plt.pie(smokers.values,labels=smokers.index,startangle=290, colors = [cmap(colors[0]),cmap(colors[1])],shadow=True,explode=(0.2,0.3),autopct='%1.1f%%')
plt.axis('equal')
legend = plt.legend(smokers.index)
legend.get_frame().set_edgecolor('r')
plt.title('Palacze')
plt.show()
```

```python
cmap = plt.cm.get_cmap('Set2')

fig,ax = plt.subplots() 
bar = ax.bar([1,2],smokers.values,color=[cmap(5),cmap(2)],yerr=(5,5))
handles, labels = ax.get_legend_handles_labels()
ax.legend(bar,('Yes','No'))
plt.grid(True,linewidth=0.5,color = '#cccccc',linestyle='--')
ax.set_xticks([1,2])
ax.set_xticklabels(smokers.index , rotation=17)
ax.set_ylim(ymax=180)

def autolabel(bar):
    for rect in bar:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')
        
autolabel(bar)
sns.despine() #getting rid of upper and right border
plt.show()
```
